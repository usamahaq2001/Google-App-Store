```python
def is_english(app_name):
    lst = []
    for i in app_name:
        if ord(i) > 127:
            lst.append(False)
        else:
            lst.append(True)
    check = set(lst)
    if False in check:
        return False
    else:
        return True
```

```python
# allow max 3 ascii in the app name as sometimes emojis are also used with english char
def is_english(app_name):
    non_ascii_count = 0

    for i in app_name:
        if ord(i) > 127:
            non_ascii_count += 1

    return non_ascii_count <= 3
```
